## Auto-updating FEC campaign funds of 2024 Presidential candidates

What is the url of the documentation? 
https://api.open.fec.gov/developers/#/candidate/get_v1_candidates_search_

```python
!pip install python-dotenv
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv('PROJECT_API_KEY')
```

```python
import requests
import pandas as pd
```

### 1. President Joseph Biden

```python
#Biden profile

url = "https://api.open.fec.gov/v1/candidates/search/"

params = {
    "api_key": api_key,
    "office": "P",
    "election_year": 2024,
    "election_type": "P",
    "name":  "BIDEN",
    "has_raised_funds": "true",
    "sort_nulls_last": "false"
}

response = requests.get(url, params=params)
biden_profile = response.json()
```

```python
biden_profile['results']
```

```python
#Campaign expenditures
url = f'https://api.open.fec.gov/v1/candidates/totals/?page=1&per_page=20&q=Biden&candidate_id=P80000722&election_year=2024&office=P&election_full=true&has_raised_funds=true&is_active_candidate=true&sort=-election_year&sort_hide_null=false&sort_null_only=false&sort_nulls_last=false&api_key={api_key}'

response = requests.get(url)
response

biden_data = response.json()
biden_data
```

```python
biden_data.keys()
```

```python
biden_data['results']
```

```python
biden_results = biden_data['results']
df_biden = pd.DataFrame(biden_results, columns=["candidate_id", "name", "party_full", "coverage_start_date", "coverage_end_date", "cash_on_hand_end_period", "individual_itemized_contributions", "other_political_committee_contributions" ])  
df_biden
```
